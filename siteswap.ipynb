```python
def jugglable(w):
    p = len(w)
    sites = []
    for i, c in enumerate(w):
        sites.append((int(c, 36) + i) % p)
    return sorted(sites) == list(range(p))
```

```python
jugglable_names = []
for p in prenoms["prenom"]:
    p = p.replace(" ", "")
    if jugglable(p):
        print(p)
        jugglable_names.append(p)
```

```python
name_list = ["elise", "emma", "frank", "gabr", "gertrudes", "ian", "jeane", "janet", "josh", "justine", "kim", "laurie", "lou", "luc", "maya", "muriel", "nadya", "paulo", "paule", "sofia", "santa", "theo"]

for n in name_list:
    b = sum([int(c, 36) for c in n]) / len(n)
    print("{} : {}".format(n, b))
```
